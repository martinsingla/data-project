# Pandas-Challenge




```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)

# The bin categories and labels are set in the original purchase data frame, so we do not need to add Age Ranges column
# in a copy of the data frame, instead we add this column to the original one and then we will work with this column further.

age_bin=[0,9,14,19,24,29,34,39,100]
age_label=["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]
purchase_data["Age Ranges"]=pd.cut(purchase_data["Age"],age_bin, labels=age_label, include_lowest=True)

purchase_data
```

## Player Count

* Display the total number of players




```python
# In order to get the real total players we require only unique Player names in SN Column

total_players=purchase_data["SN"].nunique()
pd.DataFrame({"Total Players":[purchase_data["SN"].nunique()]})
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#  len(purchase_data.groupby(["Item ID"]).count())
# Here we are doing a brief summary of the original purchase data, using nunique, mean, count, and sum to get a general
# overview of data purchase

item_count=purchase_data["Item ID"].nunique()
avg_price=purchase_data["Price"].mean()
purchase_count=purchase_data["Purchase ID"].count()
total_revenue=purchase_data["Price"].sum()

summary_df=pd.DataFrame({"Number of Unique Items":[item_count],
                         "Average Price":[avg_price],
                         "Number of Purchases":[purchase_count],
                         "Total Revenue":[total_revenue]})

summary_df["Average Price"]=summary_df["Average Price"].map("${:,.2f}".format)
summary_df["Total Revenue"]=summary_df["Total Revenue"].map("${:,.2f}".format)
summary_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# For the gender demographics we need to work with the unique players df, we create a new copy called pd_unique
# So we can group by gender and get the total stats in a new data frame called gender_df

pd_unique=purchase_data.drop_duplicates("SN")
pd_uniquebygen=pd_unique["Gender"].value_counts()

gender_df=pd.DataFrame(pd_uniquebygen)
gender_df["Percentage of Players"]=gender_df["Gender"]/total_players
gender_df["Percentage of Players"]=gender_df["Percentage of Players"].map("{:.2%}".format)
gender_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Important notice to mention from this cell, is when we want to calculate the avg purachse per Person, we need to call 2 tables
# The original one where we have the complete purchase data so we know the total amount purchased by age range and the 
# pd_unique df so we know how many unique people actually has a purchase, this is very important so we don´t consider duplicated
# people in each age range

purchasin_df=purchase_data.groupby("Gender")

purchase_count=purchasin_df["Purchase ID"].count()
average_purchase=purchasin_df["Price"].mean()
purchase_value=purchasin_df["Price"].sum()
avg_person=purchasin_df["Price"].sum()/pd_unique["Gender"].value_counts() 


purchasing_summary=pd.DataFrame({"Purchase Count":purchase_count,
                                 "Average Purchase Price":average_purchase,
                                 "Total Purchase Value":purchase_value,
                                 "Avg Total Purchase per Person": avg_person})


purchasing_summary["Average Purchase Price"]=purchasing_summary["Average Purchase Price"].map("${:,.2f}".format)
purchasing_summary["Total Purchase Value"]=purchasing_summary["Total Purchase Value"].map("${:,.2f}".format)
purchasing_summary["Avg Total Purchase per Person"]=purchasing_summary["Avg Total Purchase per Person"].map("${:,.2f}".format)

summary_df

purchasing_summary
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# With this dataframe I am colecting only the unique customers by their Age Range so we can calculate the Avg Purchase per Age Range
pd_unique_age=pd_unique.groupby(["Age Ranges"])

# Now we do the same, but using all purchases
pd_age_range=purchase_data.groupby(["Age Ranges"])


purchase_analysis=pd.DataFrame({"Purchase Count":pd_age_range["Purchase ID"].count(),
                               "Average Purchase Price":pd_age_range["Price"].mean(),
                               "Total Purchase Value":pd_age_range["Price"].sum(),
                               "Avg Total Purchase per Person":pd_age_range["Price"].sum()/pd_unique_age["Purchase ID"].count()
                               })

purchase_analysis["Average Purchase Price"]=purchase_analysis["Average Purchase Price"].map("${:,.2f}".format)
purchase_analysis["Total Purchase Value"]=purchase_analysis["Total Purchase Value"].map("${:,.2f}".format)
purchase_analysis["Avg Total Purchase per Person"]=purchase_analysis["Avg Total Purchase per Person"].map("${:,.2f}".format)
purchase_analysis
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# In this cell we are grouping the data by player, so we can analyze who are the top spenders and we can get some extra info
# of their historial purchase data

top_spender=purchase_data.groupby(["SN"])

top_spender_df=pd.DataFrame({"Purchase Count":top_spender["Purchase ID"].count(),
                             "Average Purchase Price":top_spender["Price"].mean(),
                             "Total Purchase Value":top_spender["Price"].sum()})

sort_top_spender=top_spender_df.sort_values("Total Purchase Value",ascending=False)

sort_top_spender["Average Purchase Price"]=sort_top_spender["Average Purchase Price"].map("${:,.2f}".format)
sort_top_spender["Total Purchase Value"]=sort_top_spender["Total Purchase Value"].map("${:,.2f}".format)
sort_top_spender.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame

<!--  -->

```python
# Here we are groupying by two fields, Item ID and Item Name, with this we can create a summary for the items that have
# the greastest purchase count and we can have a more accurate view of the price at what they are being sold and the total
# income from each item.

items_df=purchase_data[["Item ID","Item Name","Price"]]
items_groupby=items_df.groupby(["Item ID", "Item Name"])

items_df=pd.DataFrame({"Purchase Count":items_groupby["Item Name"].count(),
                      "Item Price":items_groupby["Price"].mean(),
                      "Total Purchase Value":items_groupby["Price"].sum()
    
})
popular_sorted=items_df.sort_values("Purchase Count", ascending=False)
popular_sorted["Item Price"]=popular_sorted["Item Price"].map("${:,.2f}".format)
popular_sorted["Total Purchase Value"]=popular_sorted["Total Purchase Value"].map("${:,.2f}".format)
popular_sorted.head()
```

```python
# Lastly, we only sort the last df by Total Purchase Value so, now we can see what are the products that have generated more
# income to the business, this gives a valuable overview of the business as we now know who is spending more, what are some
# important caracteristics of our customers (Age range that spend more), we know what are the products that are being sold
# more in quantity and the products that have had the greatest revenue.

profitable=items_df.sort_values(["Total Purchase Value"], ascending=False)
profitable.head()
```
